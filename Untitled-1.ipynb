```python
import numpy as np
import pandas as pd
import itertools

def expand_grid(data_dict):
    rows = itertools.product(*data_dict.values())
    return pd.DataFrame.from_records(rows, columns=data_dict.keys())

grr = np.linspace(0,20,20)
gruh = np.linspace(220, 290, 20)

goo = expand_grid({"angle": grr, "freq": gruh})


print(goo)

print(goo.iloc[1]["angle"])
```
